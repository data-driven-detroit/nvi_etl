```python
import pandas as pd

frame = pd.read_csv("Q:\\3_Projects\\NVI\\2025\\nvi_survey_data_2025_20260226.csv")
datadict = pd.read_csv(
    "P:\\2025_Projects\\NVI25\\Development\\Workspace\\codebook_cleanup\\nvi_codebook_organization_20260225.csv", 
    encoding='cp1252',
    keep_default_na=False
)
```

```python
COLUMN = 74

if COLUMN < 4: raise ValueError("Coded columns start at 4")
missing_cols = {16, 17}
if COLUMN in missing_cols: raise ValueError(f"Some columns are missing: {missing_cols}")

question = datadict[['label', 'topic_text', 'question_text']].drop_duplicates(subset='label').iloc[COLUMN]

if pd.notna(question["topic_text"]):
    print(question["topic_text"])

print(question["question_text"])

values = datadict[datadict['label'] == question["label"]]
texts = values[["code", "text"]].set_index("code")

(
    frame[question["label"]]
    .astype(pd.Int64Dtype())
    .value_counts(dropna=False)
    .to_frame()
    .join(texts, how="left")
    .sort_index()
    .fillna({"text": "[BLANK]"})
)
```
